## Data Cleaning

```python
import re
import pandas as pd
 
# Get the English words from each row
def get_english_words(title):
    english_words = re.findall(r"\b[a-zA-Z]+\b", title)
    if english_words:
        return " ".join(english_words)
    else:
        return "N/A"
    
def get_english_words_c(description):
    english_words = re.findall(r"\b[a-zA-Z]+\b", description)
    if english_words:
        return " ".join(english_words)
    else:
        return "N/a"
def get_english_words_usd(price):
    english_words = re.findall(r"\b[1-9-0]+\b", price)
    if english_words:
        return " ".join(english_words)
    else:
        return "N/a"

# Replace the Title column with the English words
df["Title"] = df["Title"].apply(get_english_words)
df["Description"] = df["Description"].apply(get_english_words_c)
df["Price"] = df["Price"].apply(get_english_words_usd)
# Print the DataFrame
df
```

```python
df['Price']= df['Price'].apply(lambda x: x.replace(' ' ,'')).astype(float)
```
